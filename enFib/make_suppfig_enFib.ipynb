```python
import scanpy as sc
import scFates as scf
import scvelo as scv
import matplotlib.pyplot as plt
scv.set_figure_params(figsize=(4,4),dpi_save=300)
scf.set_figure_pubready()
```

```python
adata_glia=sc.read("adata_glia.h5ad")
adata=sc.read("adata_carr.h5ad")
```

```python
scv.pl.umap(adata_glia,color="devtime")
```

```python
scv.pl.umap(adata_glia,color="milestones",title="",save="enfib_carr_ours.pdf",legend_fontoutline=False)
scv.pl.umap(adata,color="leiden",legend_loc="on data",title="",save="enfib_carr.pdf",legend_fontoutline=False)
```

```python
axs=scv.pl.umap(adata_glia,color=["Cp","Sox9","Enpp2"],ncols=4,show=False,perc=[0,99.5],layer="palantir_imp")
```

```python
axs=scv.pl.umap(adata_glia,color=["Cp","Sox9","Enpp2"],layer="palantir_imp",smooth=True,ncols=4,show=False)
```

```python
axs=scv.pl.umap(adata_glia,color=["Cp","Sox9","Enpp2"],ncols=4,smooth=True,show=False)
for ax in axs:
    ax.set_title(ax.get_title(),style="italic",size=20)
plt.savefig("./figures/scvelo_enFib_ours.pdf",dpi=300)
scv.pl.umap(adata,color=["Cp","Sox9","Enpp2"],ncols=3,smooth=True,title="",save="enFib_carr.pdf")
```
