```r
library(MouseSciaticNerve)
library(Seurat)
library(SeuratData)
library(SeuratDisk)
```

```r
dataset="InjUninjMesenchymalCombined"
filePath <- system.file(paste0(dataset,"/",dataset,".RData"),package="MouseSciaticNerve")
load(filePath)
DimPlot(InjUninjMesenchymalCombined_seur)
```

```r
InjUninjMesenchymalCombined_seur=UpdateSeuratObject(InjUninjMesenchymalCombined_seur)

InjUninjMesenchymalCombined_seur@reductions$ica=NULL

SaveH5Seurat(InjUninjMesenchymalCombined_seur, filename = "InjUninjMesenchymalCombined_seur.h5Seurat",overwrite = T)

Convert("InjUninjMesenchymalCombined_seur.h5Seurat", dest = "h5ad",overwrite = T)
```
