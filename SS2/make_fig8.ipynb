```python
import scanpy as sc
import scFates as scf
adata=sc.read("data/adata_assigned.h5ad")
```

```python
sc.pl.umap(adata,color="leiden",legend_loc="on data")
```

```python
sc.set_figure_params(dpi=150)
sel=["19","3","26","15","23","13","26","8","24","27","14","11","18","16"]
sc.pl.umap(adata,color="leiden",legend_loc="on data",groups=sel)
```

```python
adata=adata[adata.obs.leiden.isin(sel)]
```

```python
scf.set_figure_pubready()
```

```python
import scvelo as scv
scv.set_figure_params(figsize=(4,4))
```

```python
scv.pl.umap(adata,color="location",add_outline=True,s=20,legend_loc="none",title="",save="Fig7a.pdf",dpi=300)
```

```python
import matplotlib.pyplot as plt
genes=["Sox10","Plp1","Mpz","Pmp22"]
axs=scv.pl.umap(adata,color=genes,add_outline=True,s=20,layer="palantir_imp",legend_loc="none",title="",show=False)
for a,g in zip(axs,genes):
    a.annotate(g,(0.35,.7),xycoords ="axes fraction",style="italic",size=18)
plt.savefig("figures/fig7c.pdf", dpi=300)
```

```python
scv.set_figure_params(figsize=(3,4))
genes=["Ascl1","Phox2b"]
axs=scv.pl.umap(adata,color=genes,add_outline=True,s=15,layer="palantir_imp",ncols=1,legend_loc="none",title="",show=False)
for a,g in zip(axs,genes):
    a.annotate(g,(0.35,.7),xycoords ="axes fraction",style="italic",size=16)
    
plt.savefig("figures/fig7d.pdf", dpi=300)
```
